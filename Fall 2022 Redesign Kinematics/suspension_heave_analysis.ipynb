```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle
cycol = cycle('bgrcmk')
```

```python
df = pd.read_csv('/Users/janniezhong/Documents/FSAE/susp_data_heave.csv')
df = df.iloc[8:, :]

y_cols = ["Mechanical Trail [Left] [Front] (in)","Toe Angle [Left] [Front] (deg)","Toe Angle [Left] [Rear] (deg)","Camber Angle [Left] [Front] (deg)","Camber Angle [Left] [Rear] (deg)","Mechanical Trail [Left] [Rear] (in)","Lower A-Arm Chassis Fore Z [Left] [Front] (in)","Lower A-Arm Chassis Aft Z [Left] [Front] (in)","Upper A-Arm Chassis Fore Z [Left] [Front] (in)","Upper A-Arm Chassis Aft Z [Left] [Front] (in)","Lower A-Arm Chassis Fore Z [Left] [Rear] (in)","Lower A-Arm Chassis Aft Z [Left] [Rear] (in)","Upper A-Arm Chassis Fore Z [Left] [Rear] (in)","Upper A-Arm Chassis Aft Z [Left] [Rear] (in)","Tie Rod Chassis Z [Left] [Front] (in)","Tie Rod Chassis Z [Left] [Rear] (in)"]
x_col = "Motion [Heave] (in)" 
```

```python
# y = df["Mechanical Trail [Left] [Front] (in)"]
x = df["Motion [Heave] (in)"]
```

```python
def graphvHeave(dataCategory, columnNames):
    for col in columnNames:
        y = df[col]
        ax = plt.gca()
        ax.scatter(x, y, label = col, color=next(cycol))
    
    plt.xlabel("Heave (in)")
    plt.ylabel(dataCategory)
    plt.title(dataCategory + " v.s. Heave (in)")
#     plt.plot(x1, model1, color="lightcoral", label="Linear(100V)")

    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    # Put a legend to the right of the current axis
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
# graphvHeave("Mechanical Trail [Left] [Front] (in)")
# y = ["Tie Rod Chassis Z [Left] [Rear] (in)"]

toeAngle = ["Toe Angle [Left] [Front] (deg)","Toe Angle [Left] [Rear] (deg)"]
camberAngle = ["Camber Angle [Left] [Front] (deg)","Camber Angle [Left] [Rear] (deg)"]
mechanicalTrail = ["Mechanical Trail [Left] [Front] (in)","Mechanical Trail [Left] [Rear] (in)"]
frontSusp = ["Lower A-Arm Chassis Fore Z [Left] [Front] (in)","Lower A-Arm Chassis Aft Z [Left] [Front] (in)","Upper A-Arm Chassis Fore Z [Left] [Front] (in)","Upper A-Arm Chassis Aft Z [Left] [Front] (in)", "Tie Rod Chassis Z [Left] [Front] (in)"]
rearSusp = ["Lower A-Arm Chassis Fore Z [Left] [Rear] (in)","Lower A-Arm Chassis Aft Z [Left] [Rear] (in)","Upper A-Arm Chassis Fore Z [Left] [Rear] (in)","Upper A-Arm Chassis Aft Z [Left] [Rear] (in)", "Tie Rod Chassis Z [Left] [Rear] (in)"]
graphvHeave("Toe Angle (deg)", toeAngle)
graphvHeave("Camber Angle (deg)", camberAngle)
graphvHeave("Mechanical Trail (in)", mechanicalTrail)
graphvHeave("Front Suspension Z (in)", frontSusp)
graphvHeave("Rear Suspension Z (in)", rearSusp)


# for s in x_cols:
#     graphvHeave(s)
```
